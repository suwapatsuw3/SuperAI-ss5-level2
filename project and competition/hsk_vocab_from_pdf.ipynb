```python
with pdfplumber.open('/content/vocabulary_band1.pdf') as pdf:
    # Initialize an empty string to store the extracted text
    text = ""

    # Loop through each page and extract text
    for page_num in range(len(pdf.pages)):
        page = pdf.pages[page_num]
        text += page.extract_text()

# Step 5: Print the extracted text
print(text)
```

```python
b = []
a =text
for i in range(len(a)):
  if a[i] == '\n' and a[i+1]>='1' and a[i+1]<='9':
    j = i+2 #j is second letter of the line
    while a[j] != ' ': #j+1 is first char
      j += 1
    k = j+2 #k is second char
    while a[k] <'a' or a[k]>'z': #k is last char-1
      if a[k]==' ' or a[k]=='à':
        break
      k+=1
    for m in range(k+1,len(a)):
      if a[m] == '\n':
        break
    b.append([a[j+1:k],a[k+1:m]])
```

```python
each = 100
start = 9
w = b[start*each:(start+1)*each] #[:1000] [1000:2000] [2000:3000] [3000:4000] [4000:5000] [5000:]
```

```python
import random
import time
```

```python
while len(w)!=0:
  #break
  random.shuffle(w)
  r=[]
  for i in range(len(w)):
    print(w[i][0])
    time.sleep(2)
    c = input(f'{w[i][1]} : ')
    if c == ';':
      r.append(i)
    if c == 'pause':
      break
  d=0
  for v in r:
    v-=d
    w.pop(v)
    d+=1
  print('end')
  if c == 'pause':
    break
print('done')
##additional : if want to collect history
#  for j in range(len(b)):
#    if b[j][0] == w[i]:
#      b[j][1] = c
#      break
##final : can chack translation
```
